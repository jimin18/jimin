```
import numpy as np
import matplotlib.pyplot as plt
```

```
x= np.arange(-5.0, 5.0, 0.1)
```

```
x.shape
```

```
def step_function(x):
  return np.array(x>0, dtype=np.int)
```

```
y= step_function(x)
```

```
plt.plot(x, y)
plt.show()
```

$f(x) = \cfrac{1}{1+e^{-x}}$

```
def sigmoid(x):
  return 1/ (1 +np.exp(-x))
```

```
x = np.arange(-5.0, 5.0, 0.1)
y2=sigmoid(x)
```

```
plt.plot(x, y2)
plt.grid()
plt.show()
```

```
def sigmoid(x, c1=1):
  return 1/ (1 +np.exp(-c1 * x))
```

```
y2= sigmoid(x, 2.0)
```

```
plt.plot(x, y2)
plt.grid()
plt.show()
```

```
for i in range(5):
  c1 = 0.2 * i + 0.2
  y2 = sigmoid(x, c1)
  plt.plot(x, y2)
  plt.grid()
```

```
def sigmoid3(x, c1,c2):
  return 1/ (1 +np.exp(-c1 * (x-c2)))
```

```
y3= sigmoid3(10,2,x)
plt.plot(x, y3)
plt.grid()
```

```
for i in range(10):
  c2 = 0.5 * i 
  c1 = 2

  y3 = sigmoid3(x, c1, c2)
  plt.plot(x, y3)
  plt.grid()
```
